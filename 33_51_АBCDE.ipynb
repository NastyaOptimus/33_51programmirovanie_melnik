### Задание Е на стр. 35


```python
import random  # нужно для выбора случайной фразы

class Parrot:
    def __init__(self, text):
        # Создаём список фраз и сразу добавляем первую фразу
        self.phrases = [text]
    
    def say(self, repeat=1):
        # Выбираем случайную фразу из списка
        random_phrase = random.choice(self.phrases)
        # Печатаем её нужное количество раз
        print((random_phrase + " ") * repeat)
    
    def learn(self, new_phrase):
        # Добавляем новую фразу в список
        self.phrases.append(new_phrase)

# Используем
p = Parrot("Гав!")
p.say()
p.learn("Мяу!") 
p.say() 
p.say(3) 
```

### Задание А на стр. 48

```python
class LampRow:
    def __init__(self):
        # Начальное состояние - все лампочки выключены
        self.__state = "00000000"
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        if len(new_state) == 8 and all(c in "01" for c in new_state):
            self.__state = new_state
        else:
            print("Ошибка! Нужно 8 символов (только 0 и 1). Устанавливаю все нули.")
            self.__state = "00000000"
    
    def show(self):
        # Показывает лампочки: 0 -> '-', 1 -> '*'
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)

# Пример использования
lamps = LampRow()
lamps.show()                    # выведет: --------
lamps.state = "10101010"
print(lamps.state)              # выведет: 10101010
lamps.show()                    # выведет: *-*-*-*-
```

### Задание В на стр. 49

```python
class LampRow:
    def __init__(self, count=8):
        # Запоминаем количество лампочек
        self.__count = count
        # Начальное состояние - все выключены
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        # Проверяем длину строки
        if len(new_state) == self.__count:
            # Проверяем, что все символы - 0 или 1
            correct = True
            for c in new_state:
                if c != "0" and c != "1":
                    correct = False
                    break
            
            if correct:
                self.__state = new_state
            else:
                print("Ошибка! Можно использовать только 0 и 1")
                self.__state = "0" * self.__count
        else:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state = "0" * self.__count
    
    def show(self):
        # Показываем лампочки: 0 -> '-', 1 -> '*'
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            else:
                result += "*"
        print(result)

# Пример использования
lamps = LampRow(6)        # создаём ряд из 6 лампочек
lamps.show()              # выведет: ------
lamps.state = "101010"
print(lamps.state)        # выведет: 101010
lamps.show()              # выведет: *-*-*-
lamps.state = "10101010"  # ошибка! (нужно 6 символов)
print(lamps.state)        # выведет: 000000
lamps.show()              # выведет: ------
```

### Задание С на стр. 50

```python
class LampRow:
    def __init__(self, count=8):
        self.__count = count
        self.__state = "0" * count
    
    @property
    def state(self):
        return self.__state
    
    @state.setter
    def state(self, new_state):
        # Проверяем длину строки
        if len(new_state) != self.__count:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state = "0" * self.__count
            return
        
        # Проверяем, что все символы - 0, 1 или 2
        for c in new_state:
            if c not in "012":
                print("Ошибка! Можно использовать только 0, 1 или 2")
                self.__state = "0" * self.__count
                return
        
        # Если всё хорошо
        self.__state = new_state
    
    def show(self):
        # Показываем лампочки:
        # 0 -> '-' (выключена)
        # 1 -> '*' (красная)
        # 2 -> 'o' (зелёная)
        result = ""
        for lamp in self.__state:
            if lamp == "0":
                result += "-"
            elif lamp == "1":
                result += "*"
            else:  # lamp == "2"
                result += "o"
        print(result)

# Пример использования
lamps = LampRow(6)
lamps.show()                    # выведет: ------
lamps.state = "102102"
print(lamps.state)              # выведет: 102102
lamps.show()                    # выведет: *-o*-o
lamps.state = "10201010"        # ошибка! (нужно 6 символов)
print(lamps.state)              # выведет: 000000
lamps.show()                    # выведет: ------
```

### Задание D на стр. 51

```python
class LampRow:
    def __init__(self, count=8):
        self.__count = count
        # Храним состояние как целое число (0 - все выключены)
        self.__state_int = 0
    
    @property
    def state(self):
        # Преобразуем целое число в строку при чтении
        # Переводим в троичную систему и дополняем нулями слева
        result = ""
        temp = self.__state_int
        for i in range(self.__count):
            # Получаем последнюю цифру в троичной системе
            digit = temp % 3
            result = str(digit) + result
            temp //= 3
        
        return result
    
    @state.setter
    def state(self, new_state):
        # Проверяем длину строки
        if len(new_state) != self.__count:
            print(f"Ошибка! Нужно ровно {self.__count} символов")
            self.__state_int = 0
            return
        
        # Проверяем, что все символы - 0, 1 или 2
        for c in new_state:
            if c not in "012":
                print("Ошибка! Можно использовать только 0, 1 или 2")
                self.__state_int = 0
                return
        
        # Преобразуем строку в целое число (троичная система)
        value = 0
        for c in new_state:
            value = value * 3 + int(c)
        
        self.__state_int = value
    
    def show(self):
        # Получаем строку состояния через свойство state
        state_str = self.state
        
        # Показываем лампочки
        for lamp in state_str:
            if lamp == "0":
                print("-", end="")
            elif lamp == "1":
                print("*", end="")
            else:  # lamp == "2"
                print("o", end="")
        print()

# Пример использования
lamps = LampRow(6)
lamps.show()                    # ------
lamps.state = "102102"
print(lamps.state)              # 102102
lamps.show()                    # *-o*-o
lamps.state = "10201010"        # ошибка! (нужно 6 символов)
print(lamps.state)              # 000000
lamps.show()                    # ------
```
